This notebook performs Exploratory Data Analysis on the Titanic dataset.
The goal is to identify patterns and insights related to passenger survival.

Tools used:
- pandas
- matplotlib
- seaborn



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

Missing values analysis

```python
df.isnull().sum()
```

## Missing Values Insight
- Age has missing values
- Deck has many missing values
- Embarked has few missing values


```python
sns.countplot(x="survived", data=df)
plt.title("Survival Count")
plt.show()
```

Most passengers did not survive the Titanic disaster.


```python
sns.countplot(x="sex", hue="survived", data=df)
plt.title("Survival by Gender")
plt.show()
```

Females had a much higher survival rate compared to males.


```python
sns.countplot(x="class", hue="survived", data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

Passengers in 1st class had the highest survival rate.


```python
plt.figure(figsize=(8,5))
sns.histplot(df["age"], bins=30, kde=True)
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Key Insights

- Females were more likely to survive than males
- Higher-class passengers had better survival chances
- Age played a role but was not the strongest factor
- Socioeconomic status influenced survival significantly
